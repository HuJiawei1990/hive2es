```python
import pandas as pd
import numpy as np
import time
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
warning_full = pd.read_csv("C://workspace//Python//PyCharm//hive2es//data//warning.csv", 
                           header=0, sep='\t', encoding="utf-8")
warning_full.sample(10)

warning_full['warn_time'] = pd.to_datetime(warning_full['warn_time'])
```

```python
warning_tczf_ct_trans = warning_full[(warning_full.warn_topic == 'warning_tczf_ct_trans') &
                                     (warning_full.status == 1) & 
                                     (warning_full.warn_time >= "2018-01-01 00:00:00") &
                                     (warning_full.warn_time <= "2018-04-04 00:00:00")]

warning_tczf_ct_trans.head(10)
```

```python
warning_tczf_avg_cost = warning_full[(warning_full.warn_topic == 'warning_tczf_avg_cost') &
                                     (warning_full.status == 1) & 
                                     (warning_full.warn_time >= "2018-01-01 00:00:00") & 
                                     (warning_full.warn_time <= "2018-04-04 00:00:00")]


warning_tczf_avg_cost.head(10)
```

```python
warning_tczf_rate_success = warning_full[(warning_full.warn_topic == 'warning_tczf_rate_success') &
                                     (warning_full.status == 1) & 
                                     (warning_full.warn_time >= "2018-01-01 00:00:00") & 
                                     (warning_full.warn_time <= "2018-04-04 00:00:00")]

pd.to_datetime(warning_tczf_rate_success.warn_time)

warning_tczf_rate_success.head(10)
```

```python
tczf_value = pd.read_csv("C://workspace//Python//PyCharm//hive2es//output//tczf", 
                           header=0, sep=',', encoding="utf-8")

tczf_value.sample(10)
```

```python
tczf_value['datex_time'] = pd.to_datetime([datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
                                          for ts in tczf_value["timestamp"]])
```

```python
for item in tczf_value["timestamp"]:
    print(item)
    break
```

```python
tczf_value.sample(5)
```

```python
warning_tczf_rate_success.index = warning_tczf_rate_success.warn_time
warning_tczf_avg_cost.index = warning_tczf_avg_cost.warn_time
warning_tczf_ct_trans.index = warning_tczf_ct_trans.warn_time

```

```python
tczf_value.index = tczf_value.datex_time
```

```python
date_ct = warning_tczf_ct_trans['id'].groupby(warning_tczf_ct_trans.index.map(lambda x: datetime.date(x))).count()
for idx, c2 in enumerate(date_ct):
    
    print("%s, %s" % (date_ct.index[idx], str(c2))) 
```

```python
warning_tczf_rate_success['id'].groupby(warning_tczf_rate_success.index.map(lambda x: datetime.date(x))).count() 
```

```python
date1 = '2018-03-30'
#warning_tczf_avg_cost[date1]['index']
plt.figure(figsize=[15, 6])

max_value = max(tczf_value[date1]['avg_cost'])

for ts in warning_tczf_avg_cost[date1].index:
    plt.axvline(ts, color='lightcoral')
plt.plot(tczf_value[date1].index, tczf_value[date1]['avg_cost'], color='b')

plt.ylim([0, max_value + 1])

plt.show()
```

```python
date1 = '2018-03-03'
plt.figure(figsize=[20, 4])

min_value = min(tczf_value[date1]['rate_success'])

for ts in warning_tczf_rate_success[date1].index:
    plt.axvline(ts, color='lightcoral')
plt.plot(tczf_value[date1].index, tczf_value[date1]['rate_success'], color='b')

plt.ylim([min_value - 5, 100])

plt.show()
```

```python
date1 = '2018-02-12'
plt.figure(figsize=[15, 6])

max_value = max(tczf_value[date1]['ct_trans'])

for ts in warning_tczf_ct_trans[date1].index:
    plt.axvline(ts, color='lightcoral')
plt.plot(tczf_value[date1].index, tczf_value[date1]['ct_trans'], color='b')

plt.ylim([0, max_value * 1.1])

plt.show()
```
